###Which one should you use? (pyplot vs Matplotlib Object Oriented Method?)
*When plotting something quickly, okay to use the pyplot method
*When plotting something more advanced, use the Object Oriented Method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50.plot(kind = "scatter",
             x="age",
             y="chol",
             c="target");
##This is the pyPlot method
```

```python
##We now mak use of the Object Oriented method
##OO method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
#ax.set_xlim([45, 100])
```

```python
##Object-Oriented Method From Scratch
fig, ax = plt.subplots(figsize=(10,6))

#Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

##customize the plot
ax.set(title="Heart Disease and Cholestrol levels",
       xlabel="Age",
       ylabel="Cholesterol")

##add a legend
ax.legend(*scatter.legend_elements(), title="Target");

##add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
##We are back, now lets complete the final part of matplotlib
```

```python
##Subplot of cho, age, thalach
```

```python
over_50.head()
```

```python
##Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                              sharex=True)

##Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

##Customize ax0
ax0.set(title="Heart Disease and cholesterol levels",
    ##    xlabel="Age", commented out because sharex=True
        ylabel="Cholesterol")
##Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

##add a Mean Line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

##SECOND SCATTER
##add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])
##Customize ax1
ax1.set(title="Heart Disease and Max heart rate",
        xlabel="Age",
        ylabel="Max Heart Rate");

##add a legend 
ax1.legend(*scatter.legend_elements(), title="Target")

##Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

##Add a title to the final full figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");  


##This figure was built using OBJECT-ORIENTED-OPTIMIZATION
```
