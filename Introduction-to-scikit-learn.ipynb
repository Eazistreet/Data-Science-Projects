```python
import pandas as pd
import numpy as np
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
##Create x-----> which is called the features matrix
X = heart_disease.drop("target", axis=1)

##Create y---->labels
y = heart_disease["target"]
```

```python
##2. Choose the right model and hyperparameters(dials on a model that you can tune to make the model better or worse)
from sklearn.ensemble import RandomForestClassifier ##This is a classification machine learning model capable of learning patterns in data and then classifying whether a sample/row is particularly something
clf = RandomForestClassifier()

##We will keep the default hypaparameters
clf.get_params()
```

```python
##3. Fit the model to the training data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) ##This line splits our data into a set of x-training data, y-training data as well as corrspondondig test data
                                                    ##0.2 splits the data with a ration of 20-80% ie 80% training data , 20% test data
```

```python
clf.fit(X_train, y_train); ##Find patterns in training data`
```

```python
X_train
```

```python
##now we make a prediction
y_label = clf.predict(np.array([0,2,3,4]))

```

```python
X_test
```

```python
y_preds = clf.predict(X_test)
y_preds
##Prints out the possible valurs of our y_predictions
```

```python
y_test
```

```python
##4. evaluate the model on the training data and test data
clf.score(X_train, y_train) ##Module poduces a 1 which is 100% accuracy, the results from the fact that the mdoel had full access to the data as well as the names of the train data set
```

```python
clf.score(X_test, y_test) ##the model scored a 0.78 accuracy because it does not have access to the data names etc
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_preds)) ##Produces some classification metrics that compare the test labels to the prediction labels we made with our model
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds) ##WE WANT TO NOW IMPROVE OUR MODEL ie >0.78
```

```python
#5. Improve our model
##Try different amount of n_estimators
np.random.seed(42)
for i in range(10,100,10):
    print(f"trying to model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model Accuracy on test set: {clf.score(X_test, y_test)*100:2f}%")
    print("")
```

```python
##Model 40estimators is the most efficient model, we got this by adjusting one of our hyparameters
```

```python
##6. Save a model and load it
import pickle ##

pickle.dump(clf, open("random_forest_model_1.pk1", "wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_1.pk1", "rb"))
loaded_model.score(X_test, y_test)
```
